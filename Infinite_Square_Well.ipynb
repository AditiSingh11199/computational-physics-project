# Infinite Square Well

This notebook explores the analytical solutions of the one-dimensional infinite square well.

The objectives are:

- Visualise the probability density of stationary states
- Investigate the effect of the quantum number n
- Calculate and plot the quantised energy levels

# Probability Density

The stationary-state wavefunctions of the infinite square well are given by

$$
\psi_n(x)=\sqrt{\frac{2}{L}}\sin\left(\frac{n\pi x}{L}\right)
$$

The probability density is 

$$
|\psi_n(x)|^2
$$

which gives the probability of finding the particle at position x.

The following code plots the probability density for several quantum states.

```python
import numpy as np 
import matplotlib.pyplot as plt

## Parameters

L = 1.0 
x = np.linspace(0, L, 1000)

## Wavefunctions for the first three energy levels in an infinite potential well 

psi1 = np.sqrt(2/L) * np.sin(1 * np.pi * x / L)
psi2 = np.sqrt(2/L) * np.sin(2 * np.pi * x / L)
psi3 = np.sqrt(2/L) * np.sin(3 * np.pi * x / L)

## Plotting the probability densities 

fig, axs = plt.subplots(3, 1, figsize=(6,9))

psi = [psi1, psi2, psi3]

for i in range(3):
    axs[i].plot(x, psi[i]**2)
    axs[i].set_title(f"n = {i+1}")
    axs[i].set_xlabel("Position (x)")
    axs[i].set_ylabel(r"$|\psi|^2$")
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```

# Quantised Energy Levels

The allowed energy levels of a particle in an infinite square well are

$$
E_n=\frac{n^2\pi^2\hbar^2}{2mL^2}
$$

Unlike classical systems, the particle can only possess discrete energy values determined by the quantum number n.

The following code calculates and visualises the first few allowed energy levels.

```python
import matplotlib.pyplot as plt

import numpy as np

hbar = 1
m = 1
L = 1

n = np.array([1, 2, 3, 4, 5])
E = (n**2 * np.pi**2 * hbar**2) / (2 * m * L**2)

print(E)

plt.plot(n, E, marker='x')
plt.xlabel("n")
plt.ylabel("Energy")
plt.title("Infinite Square Well Energy Levels")
plt.show()
```

# Conclusions

The infinite square well demonstrates two fundamental features of quantum mechanics:

- The wavefunction forms standing-wave patterns inside the well.
- The energy spectrum is quantised, with energy increasing as n².
